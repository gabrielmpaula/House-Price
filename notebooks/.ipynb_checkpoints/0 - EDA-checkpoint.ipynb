## Loading Data 

```python
DATA_PATH = '../data/'
train_data = pd.read_csv(DATA_PATH + 'train.csv')
```

```python
train_data.head()
```

```python
target = train_data['SalePrice']
train_data = train_data.drop(['Id','SalePrice'], axis=1)
```

## Target analysis

```python
target.describe()
```

```python
plt.figure(figsize=(10,7))
sns.distplot(target)
```

Target is skewed to the right and has some outliers above 70000

```python
plt.figure(figsize=(10,7))
sns.distplot(np.log(target))
```

The competition uses the RMSLE (Root Mean Square Logaritic Error). Taking the log of the target then solves two problems. The skewness, aproximating it to a normal distribution, and now the RMSE can direct applied to the target and we'll get the RMSLE.

```python
numerical_columns = train_data.select_dtypes(['int64','float64']).columns
categorical_columns = train_data.select_dtypes(['object']).columns
print('Numerical columns: {} \nCategorical columns: {}'.format(len(numerical_columns), len(categorical_columns)))
```

## Distribution of numerical features 

```python
fig, axes = plt.subplots(nrows=18, ncols=2, figsize=(15,100))
for i, column in enumerate(numerical_columns):
    sns.distplot(train_data[column].dropna(),ax=axes[i//2,i%2])
```

## Correation between features and target 

```python
num_data = pd.concat([train_data[numerical_columns], target], axis=1)
corr = num_data.dropna().corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr)
```

```python
hm_cols = corr.nlargest(10, 'SalePrice').index
corr = np.corrcoef(num_data[hm_cols].values.T)
plt.figure(figsize=(7,5))
sns.heatmap(corr, yticklabels=hm_cols, xticklabels=hm_cols, annot=True)
```

## Cardinality of categorical features

```python
high_cardinality_features = train_data[categorical_columns].nunique().loc[train_data[categorical_columns].nunique() > 10]
train_data[categorical_columns].nunique().sort_values()
```

```python
print('Features with high cardinality: ')
print(high_cardinality_features.index.tolist())
```
